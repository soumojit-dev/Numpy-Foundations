```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
                       [1, 150000, 180000, 200000, 250000], # Arsalan
                       [2, 120000, 140000, 160000, 190000], # Chowman
                       [3, 200000, 230000, 260000, 300000], # Dominos
                       [4, 180000, 210000, 240000, 270000], # Burger King
                       [5, 160000, 180000, 205000, 230000] # BTech Chaiwala
])

print("-----SWIGGY SALES ANALYSIS-----")
print("Sales Data Shape",sales_data.shape)
print("\nData of First Three Restaurant:\n",sales_data[:3])

# Total Sales
print("\nTotal Sales Per Year of the 5 Restaurants Collectively:\n",np.sum(sales_data[:, 1:], axis=0)) #[all rows:, col_start:col_end]
print("\nTotal Sales for Year(2021-2024) of each of the 5 Restaurants:\n",np.sum(sales_data[:, 1:], axis=1))

# Minimum Sales Per Restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("\nMinimum Sales Per Restaurant:\n",min_sales)

# Maximum Sales Per Restaurant
max_sales = np.max(sales_data[:, 1:], axis=1)
print("\nMaximum Sales Per Restaurant:\n",max_sales)

# Average Sales Per Restaurant
avg_sales = np.average(sales_data[:, 1:], axis=1)
print("\nAverage Sales Per Restaurant:\n",avg_sales)

# Cumulative Sales Per Restaurant & Plotting
cum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("\nCumulative Sales(2021-2024) of each Restaurants:\n",cum_sales)

# Data Extraction
rest_id = sales_data[:,0]
years = ['2021', '2022', '2023', '2024']

# Data For Plotting
rest_ids = []
years_ = []
sales_list = []

for i in range(len(rest_id)):
    for j in range(len(years)):
        rest_ids.append(rest_id[i])
        years_.append(years[j])
        sales_list.append(str(sales_data[i,j+1]))

# Plotting
pl.figure(figsize=(10,5))

pl.plot(rest_id, min_sales, marker='s', label='Minimum Sales')
pl.plot(rest_id, max_sales, marker='<', label='Maximum Sales')
pl.plot(rest_id, avg_sales, marker='o', label='Average Sales')
pl.plot(rest_id, cum_sales, marker='>', label='Cumulative Sales')
pl.xlabel('Restaurant ID')
pl.ylabel('Sales')
pl.title('Sales Metrics Per Restaurant (2021-2024)')
pl.grid(True)
pl.show()
```
