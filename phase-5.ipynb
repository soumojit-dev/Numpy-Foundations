# Storing Image in Matrix and Converting Image to Dark Mode

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.random.rand(3,3)
arr3 = np.zeros((3,3))
```

```python
np.save('arrar1.npy',arr1)
np.save('arrar2.npy',arr2)
np.save('arrar3.npy',arr3)
```

```python
loaded_array = np.load('arrar1.npy')
print(loaded_array)
```

# NOTE
# Image data range issue:

# If the image array logo contains values between 0 and 255 (common for PNG files with dtype=uint8), then doing 1 - logo has almost no effect because:

# Example: 1 - 255 = -254, but it gets clipped or wrapped by the data type.

# If the array contains values between 0 and 1 (float format), then 1 - logo works as expected.

# You applied inversion but the values didn’t change visually because either:

# The array is in uint8 format (0–255), or

# The image is nearly symmetric, or only slight inversion was possible.
